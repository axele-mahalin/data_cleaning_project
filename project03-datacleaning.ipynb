# Importing data

```python
import pandas as pd
import numpy as np
import re
```

```python
df_og = pd.read_csv('6 - Primary Education.csv')
```

```python
df = df_og.copy()
```

```python
df_og
```

```python
df.head()
```

```python
df.describe()
```

```python
# drop useless columns
df.drop(['Sub-region','ISO3'], axis=1, inplace=True)
```

```python
# drop rows with null data 
df = df.dropna()
```

```python
# filter only the rows for which the data is over a single year
# delete rows where len(year) != 4

df["Time period"] = df["Time period"].astype('str')

df = df[df["Time period"].apply(len) ==4] 

df
```

```python
# drop outliers value


# Outliers in Time Period
df = df[df["Time period"] != "2562"]

df = df[df["Time period"] != "2076"]

#Keep only indexes that are between 0 and 1
df = df[df["Total"] <= 1]

df = df[df["Residence Rural"] <= 1] 

df = df[df["Residence Urban"] <= 1] 

df = df[df["Wealth quintile Poorest"] <= 1] 

df = df[df["Wealth quintile Richest"] <= 1] 


df
```

```python
# Correct index

df = df.reset_index(drop=True)

df
```

```python
df.to_csv('primary_education_clean.csv', sep=',', encoding="utf8")
```
